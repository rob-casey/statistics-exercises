A bank found that the average number of cars waiting during the noon hour at a drive-up window follows a Poisson distribution with a mean of 2 cars. 

Make a chart of this distribution and answer these questions concerning the probability of cars waiting at the drive-up window.

What is the probability that no cars drive up in the noon hour?

What is the probability that 3 or more cars come through the drive through?

How likely is it that the drive through gets at least 1 car?

```python
λ = 2
bank_window = stats.poisson(λ)
```

```python
x = np.arange(0, 10)
y = stats.poisson(λ).pmf(x)
plt.bar(x, y)
plt.show()
```

```python
# alt. solution

x = np.arange(0, 24, 0.25)
y = [bank_window.pmf(i) for i in x]
plt.plot(x, y)
plt.show()
```

What is the probability that no cars drive up in the noon hour?

```python
stats.poisson(2).pmf(0)
```

What is the probability that 3 or more cars come through the drive through?

```python
stats.poisson(2).sf(2)
```

How likely is it that the drive through gets at least 1 car?

```python
stats.poisson(2).sf(0)
```

```python
mean_students = 3.0
std_dev_students = 0.3

student_dist = stats.norm(mean_students, std_dev_students)
```

What grade point average is required to be in the top 5% of the graduating class?

```python
top_5 = student_dist.isf(0.5)
top_5
```

What GPA constitutes the bottom 15% of the class?

```python
bottom_15 = student_dist.ppf(0.15)
bottom_15
```

An eccentric alumnus left scholarship money for students in the third decile from the bottom of their class. Determine the range of the third decile. 

```python
student_dist.ppf([.2, .3])
```

Would a student with a 2.8 grade point average qualify for this scholarship?

```python
student_dist.cdf(2.8).round(3)

# Yes, a student with a 2.8 grade point average would quality because
# 2.8 = bottom 25% and falls within the third decile, 20% - 30% from the bottom.
```

If I have a GPA of 3.5, what percentile am I in?

```python
student_dist.cdf(3.5)

# If you have a 3.5 GPA, you're in the 95th percentile.
```

A marketing website has an average click-through rate of 2%. 

One day they observe 4326 visitors and 97 click-throughs. 

How likely is it that this many people or more click through?

```python
rate = 0.02
visits = 4326
clicks = 97

click_through = stats.binom(visits, rate).sf(clicks - 1).round(3) # survival = exclusive
click_through
```
